```python
Q2.
ANS-
The Cumulative Density Function (CDF) is a function that describes the cumulative probability of a random variable taking on a value less than or equal to a given value. 
It gives information about the probability distribution of the random variable and provides a cumulative perspective.

The CDF is denoted as F(x), where x represents a specific value of the random variable. 

The CDF satisfies three properties:-
non-decreasing (the function increases or stays constant as x increases), right-continuous (the function has no jumps), and approaches 0 as x approaches negative infinity and 1 as x approaches positive infinity.

The CDF is used for several reasons:-

1.Calculation of probabilities:-
The CDF allows for the calculation of probabilities associated with the random variable. 
For example, to find the probability of the random variable being less than or equal to a specific value x, you can evaluate the CDF at that value: P(X ≤ x) = F(x).

2.Identification of percentiles:-
The CDF provides a way to determine the percentiles of a distribution.
By finding the value of x for which F(x) equals a specific probability, you can determine the value below which that probability of the random variable falls.

3.Comparison of random variables:-
The CDF allows for the comparison of different random variables or distributions.
By comparing their CDFs, you can assess which one has higher or lower probabilities for specific values or ranges.

Example:
    
Consider a fair six-sided die. The CDF for this die can be expressed as follows:

F(x) = P(X ≤ x)

For each value of x, the CDF assigns the cumulative probability of obtaining a value less than or equal to x. For this fair dice, the CDF would be:

F(1) = 1/6 (probability of getting 1 or less)
F(2) = 2/6 (probability of getting 2 or less)
F(3) = 3/6 (probability of getting 3 or less)
F(4) = 4/6 (probability of getting 4 or less)
F(5) = 5/6 (probability of getting 5 or less)
F(6) = 6/6 = 1 (probability of getting 6 or less)

The CDF provides a cumulative perspective by showing how the probabilities accumulate as we move from smaller to larger values. 
It enables the calculation of probabilities and the understanding of the distribution's behavior across the entire range of possible values.
```

```python
Q.5
ANS-
The Bernoulli distribution is a discrete probability distribution that models a binary outcome, such as success or failure, in a single trial or experiment.
It is named after the Swiss mathematician Jacob Bernoulli. 
The distribution has a single parameter, often denoted as "p," which represents the probability of success.

The Bernoulli distribution can be used to model situations where there are only two possible outcomes, and each trial is independent of the others. 
The random variable in the Bernoulli distribution takes the value 1 with probability p (representing success) and the value 0 with probability 1-p (representing failure).

Example:- Tossing a fair coin can be modeled using a Bernoulli distribution. 
Let's say we define success as getting heads and failure as getting tails. In this case, the probability of success (getting heads) is 0.5, and the probability of failure (getting tails) is also 0.5. So, we can use a Bernoulli distribution with p = 0.5 to model this situation.

difference between the Bernoulli distribution and the Binomial distribution-

Bernoulli Distribution:-
The Bernoulli distribution models a single trial or experiment with a binary outcome (success or failure). 
It has a single parameter, p, representing the probability of success.
The random variable in the Bernoulli distribution takes the value 1 with probability p and the value 0 with probability 1-p.

Binomial Distribution:-
The Binomial distribution extends the Bernoulli distribution to multiple trials or experiments.
It models the number of successes in a fixed number of independent Bernoulli trials. 
The Binomial distribution has two parameters: n, the number of trials, and p, the probability of success in each trial.

Example:-
Suppose we conduct 10 independent trials of flipping a fair coin and count the number of heads. 
The outcome of each trial follows a Bernoulli distribution with p = 0.5. 
The total number of heads obtained in the 10 trials can be modeled using a Binomial distribution with parameters n = 10 and p = 0.5.
```

```python
Q9.
ANS-

The Central Limit Theorem (CLT) is a fundamental concept in statistics that states that, under certain conditions, the sampling distribution of the mean of a random sample approaches a normal distribution as the sample size increases, regardless of the shape of the population distribution.

Key points about the Central Limit Theorem:-

Sample Size:-
-
The Central Limit Theorem applies when the sample size is sufficiently large. There is no specific threshold for what constitutes a "sufficiently large" sample size, but as a general guideline, a sample size of 30 or greater is often considered adequate.

Independence:-
The individual observations in the sample should be independent of each other. This means that each observation should be unrelated and not influenced by the other observations.

Random Sampling:-
The samples should be selected randomly from the population of interest. Random sampling ensures that the sample is representative of the population and avoids bias.

significance of the Central Limit Theorem:-

1.Approximation:-
The Central Limit Theorem allows us to approximate the distribution of sample means as a normal distribution, even if the population distribution is not normal. 
This approximation is widely used in statistical inference, hypothesis testing, and confidence interval estimation.

2.Statistical Inference:-
The Central Limit Theorem forms the foundation for many statistical inference procedures. 
It enables us to make inferences about population parameters (such as means and proportions) based on sample statistics. 

example- 
it allows us to estimate a population mean and determine confidence intervals for the population mean.

3.Population Estimation:-
The Central Limit Theorem provides a powerful tool for estimating population parameters. 
By knowing the properties of the sampling distribution of the mean, we can make accurate predictions and draw conclusions about the population even with limited sample information.

4.Hypothesis Testing:-
The Central Limit Theorem is crucial in hypothesis testing, particularly for large sample sizes. 
It allows us to compare sample statistics (such as means or proportions) to hypothesized values and determine the statistical significance of the results.

5.Real-world Applications:-
The Central Limit Theorem has broad applications in various fields, including finance, quality control, social sciences, and natural sciences. 
It provides a basis for understanding and analyzing data when assumptions of normality are required or when approximations are necessary.
```
