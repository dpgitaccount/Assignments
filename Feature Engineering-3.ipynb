```python
Q1. What is Min-Max scaling, and how is it used in data preprocessing? Provide an example to illustrate its
application.

ANS-

Min-Max scaling, also known as normalization, is a data preprocessing technique used to scale numerical features in a dataset to a specific range. 
This range is typically defined as [0, 1], where the minimum value of the feature is scaled to 0, and the maximum value is scaled to 1. This transformation helps bring all features onto a common scale, preventing features with larger magnitudes from dominating the learning algorithms that rely on distance measures or gradient-based optimization.

Example -


from sklearn.preprocessing import MinMaxScaler

# exam scores
exam_scores = [60, 75, 90, 100]

exam_scores = [[score] for score in exam_scores]

# MinMaxScaler
scaler = MinMaxScaler()

normalized_scores = scaler.fit_transform(exam_scores)

print("Normalized scores:", normalized_scores)
```

```python
2.What is the Unit Vector technique in feature scaling, and how does it differ from Min-Max scaling?
Provide an example to illustrate its application.

ANS-

The Unit Vector technique, also known as vector normalization, is another feature scaling method used to transform numerical features in a dataset. 
Unlike Min-Max scaling, which scales features to a specific range (e.g., [0, 1]), Unit Vector scaling scales each feature vector to have a magnitude of 1 (i.e., it transforms features into unit vectors).
This scaling technique is particularly useful when you want to emphasize the direction of the data rather than its magnitude.
It's commonly used in machine learning algorithms that rely on distance measures, like K-nearest neighbors (KNN) and Support Vector Machines (SVM), where the distance between data points matters more than their magnitudes.

EXAMPLE -

import numpy as np

# Original feature vectors
features = np.array([[3, 1],
                     [0, 4],
                     [2, 2]])

norms = np.linalg.norm(features, axis=1)
unit_vectors = features / norms.reshape(-1, 1)

print("Original features:\n", features)
print("Unit vectors:\n", unit_vectors)
```

```python
Q3. What is PCA (Principle Component Analysis), and how is it used in dimensionality reduction? Provide an
example to illustrate its application.

ANS-

Principal Component Analysis (PCA) is a dimensionality reduction technique used to transform high-dimensional data into a lower-dimensional representation while preserving as much of the original data's variance as possible. 
It does so by identifying the principal components, which are linear combinations of the original features that capture the most significant patterns of variation in the data. 
The first principal component accounts for the most variance, the second principal component accounts for the second most variance, and so on.

PCA is commonly used to simplify complex datasets, remove noise, and speed up machine learning algorithms by reducing the number of features while retaining the most important information.

EXAMPLE -

import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

np.random.seed(42)
n_samples = 100
n_features = 2
X = np.random.rand(n_samples, n_features)

# PCA with 1 principal component
pca = PCA(n_components=1)
X_reduced = pca.fit_transform(X)

X_projected = pca.inverse_transform(X_reduced)

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], label='Original Data', alpha=0.5)
plt.scatter(X_projected[:, 0], X_projected[:, 1], label='Reduced Data', alpha=0.5)
plt.legend()
plt.title('PCA for Dimensionality Reduction')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
Q4. What is the relationship between PCA and Feature Extraction, and how can PCA be used for Feature
Extraction? Provide an example to illustrate this concept.

ANS-


PCA (Principal Component Analysis) and feature extraction are related concepts in the context of dimensionality reduction and data preprocessing.
Feature extraction refers to the process of transforming the original features of a dataset into a new set of features (often of lower dimension) that capture the most relevant information for a given task. 
PCA can be used as a technique for feature extraction, particularly when the original features are highly correlated.

PCA performs dimensionality reduction by finding a new set of orthogonal (uncorrelated) features called principal components, which are linear combinations of the original features. 
These principal components are ordered by the amount of variance they explain in the data.
By selecting a subset of the top principal components, you can effectively extract the most important information from the original features while reducing the dimensionality of the dataset.

EXAMPLE-

import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

# Using Iris data set
iris = load_iris()
X = iris.data

# PCA for feature extraction
pca = PCA(n_components=2)  # Reduce to 2 dimensions
X_extracted = pca.fit_transform(X)

explained_variances = pca.explained_variance_ratio_
print("Explained variances:", explained_variances)

print("Extracted features:\n", X_extracted[:10])
```

```python
# Q7. For a dataset containing the following values: [1, 5, 10, 15, 20], perform Min-Max scaling to transform the
# values to a range of -1 to 1.

# ANS-

import numpy as np

# Given dataset
data = np.array([1, 5, 10, 15, 20])

a = -1
b = 1

# Calculate min and max.
data_min = np.min(data)
data_max = np.max(data)

# Min-Max scaling formula
scaled_data = a + ((data - data_min) * (b - a)) / (data_max - data_min)

print("Original data:", data)
print("Scaled data:", scaled_data)
```

```python
Q8. For a dataset containing the following features: [height, weight, age, gender, blood pressure], perform
Feature Extraction using PCA. How many principal components would you choose to retain, and why?

ANS-


Some principal components to retain:-

Calculate Explained Variance Ratios:--
After applying PCA, the explained variance ratio for each principal component will tell you how much variance is explained by that component. You can obtain these ratios using the explained_variance_ratio_ attribute of the PCA object.

Cumulative Explained Variance:--
Calculate the cumulative explained variance by summing up the explained variance ratios. This will show you how much total variance is retained as you include more principal components.

Choose the Number of Components:--
Based on the cumulative explained variance, you can decide how many principal components to retain. A common guideline is to choose the number of components that collectively capture a high percentage of the total variance, such as 95% or 99%. This ensures that you retain most of the important information while reducing the dimensionality.



import numpy as np
from sklearn.decomposition import PCA

# Create a random dataset for illustration
np.random.seed(42)
n_samples = 100
n_features = 5
data = np.random.rand(n_samples, n_features)

# Apply PCA
pca = PCA(n_components=n_features)  # Retain all components
data_pca = pca.fit_transform(data)

# Calculation of explained variance ratios and cumulative explained variance
explained_variances = pca.explained_variance_ratio_
cumulative_explained_variance = np.cumsum(explained_variances)

# Print of explained variance ratios and cumulative explained variance
print("Explained variance ratios:", explained_variances)
print("Cumulative explained variance:", cumulative_explained_variance)
```
