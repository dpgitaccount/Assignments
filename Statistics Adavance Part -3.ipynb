```python
# Q2.
# ANS-

import math

def estimate_population_mean(sample_mean, sample_std_dev, sample_size):
    
    # Calculate the standard error
    standard_error = sample_std_dev / math.sqrt(sample_size)
    
    
    confidence_level = 0.95
    z_value = 1.96
    margin_of_error = z_value * standard_error
    
    # Calculate the lower and upper bounds of the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    
    # Return the estimated population mean and the confidence interval
    return sample_mean, (lower_bound, upper_bound)

sample_mean = 50.0
sample_std_dev = 5.0
sample_size = 100

estimated_mean, confidence_interval = estimate_population_mean(sample_mean, sample_std_dev, sample_size)

print("Estimated Population Mean:", estimated_mean)
print("Confidence Interval:", confidence_interval)

```

```python
# Q5.
# ANS-
import scipy.stats as stats

def hypothesis_test(sample1, sample2, alpha):
    # independent t-test assuming unequal variances
    t_statistic, p_value = stats.ttest_ind(sample1, sample2, equal_var=False)

    # Compare of p-value 
    if p_value < alpha:
        print("Reject the null hypothesis. There is evidence of a significant difference.")
    else:
        print("Fail to reject the null hypothesis. There is no significant difference.")

    
    return t_statistic, p_value


sample1 = [1.5, 2.0, 2.2, 1.8, 1.9]
sample2 = [1.2, 1.6, 1.8, 1.5, 2.1]
alpha = 0.05

t_statistic, p_value = hypothesis_test(sample1, sample2, alpha)

print("T-statistic:", t_statistic)
print("P-value:", p_value)
```

```python
Q8.
ANS-
The p-value is a statistical measure used in hypothesis testing to quantify the strength of evidence against the null hypothesis. 
It represents the probability of obtaining a test statistic as extreme as, or more extreme than, the observed value, assuming the null hypothesis is true. In simpler terms, the p-value measures the likelihood of observing the data or a more extreme result if the null hypothesis is actually true.

The significance of the p-value in hypothesis testing lies in its role in decision-making.

If p-value < Œ±:-
If the p-value is less than the chosen significance level (Œ±), typically 0.05 or 0.01, it indicates that the observed data is unlikely to occur under the assumption that the null hypothesis is true. 
In this case, there is sufficient evidence to reject the null hypothesis in favor of the alternative hypothesis.
The result is considered statistically significant, suggesting that there is a meaningful effect or difference.

If p-value ‚â• Œ±:-
If the p-value is greater than or equal to the significance level (Œ±), it suggests that the observed data is reasonably likely to occur by chance even if the null hypothesis is true. 
In this case, there is not enough evidence to reject the null hypothesis. The result is considered statistically non-significant, indicating that any observed effect or difference may be due to random variability or other factors, rather than a true population effect.
```

```python
Q11.
ANS-

Student's t-distribution, often referred to as the t-distribution, is a probability distribution that arises in statistical inference. It is used when the sample size is small and the population standard deviation is unknown. The t-distribution is similar to the standard normal distribution (Z-distribution), but it has heavier tails, allowing for more uncertainty in the estimates.

The t-distribution is characterized by its degrees of freedom (df). The degrees of freedom determine the shape of the distribution. As the degrees of freedom increase, the t-distribution approaches the shape of the standard normal distribution.

The t-distribution is commonly used in the following situations:

Small sample sizes:-
When working with small samples (typically less than 30), the t-distribution is used to estimate population parameters, such as the mean, when the population standard deviation is unknown. This is in contrast to the standard normal distribution, which is used for large sample sizes (typically greater than 30) or when the population standard deviation is known.

Confidence intervals:-
The t-distribution is used to construct confidence intervals for population parameters when the sample size is small and the population standard deviation is unknown. The wider confidence intervals account for the increased uncertainty associated with smaller sample sizes.

Hypothesis testing:-
In hypothesis testing, the t-distribution is used when testing hypotheses about population means or differences between means, especially when the sample size is small and the population standard deviation is unknown. It is employed to calculate the test statistic and determine the p-value, which helps in drawing conclusions about the hypotheses.

Comparison of means:-
When comparing means between two groups (e.g., treatment vs. control), the t-distribution is used to perform statistical tests, such as the independent samples t-test or paired samples t-test.
```
