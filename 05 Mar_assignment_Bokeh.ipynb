```python
Q2.
ANS-

In Bokeh, glyphs are visual markers or shapes that represent your data on a plot. 
They can be used to create various types of plots, such as scatter plots, line plots, bar charts, and more. 
Glyphs allow you to customize the appearance and attributes of the markers, such as size, color, transparency, and shape, to convey information effectively.


To add glyphs to a Bokeh plot, we are typically use the figure object and its methods to create the desired plot.

Commonly used glyphs methods are:-

circle(): Adds circular markers to the plot.
square(): Adds square markers to the plot.
line(): Adds a line to the plot.
patch(): Adds a filled area to the plot.
rect(): Adds rectangles to the plot.
image(): Adds images to the plot.

EXAMPLE- 
from bokeh.plotting import figure, show

# Create data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a new figure
p = figure(title='My Bokeh Plot', x_axis_label='X', y_axis_label='Y')

# Add glyphs to the figure
p.circle(x, y, size=10, color='blue', alpha=0.8)
p.line(x, y, line_width=2, color='red')

# Show the plot
show(p)

```

```python
Q3.
ANS-

In Bokeh, we can customize the appearance of a plot, including the axes, title, and legend, using various attributes and properties provided by the figure object and other related objects.

1. Plot Title:-
we can set the title of the plot using the title attribute of the figure object.

example - 

p = figure(title='My Bokeh Plot')


2. Axes Labels:-
we can set labels for the x-axis and y-axis using the x_axis_label and y_axis_label attributes of the figure object. 

example -

p = figure(x_axis_label='X', y_axis_label='Y')


3. Legend:-
To add a legend to your plot, we need to assign a unique legend_label attribute to each glyph you add. 
Then, we can enable the legend using the legend attribute of the figure object.

example- 

p.circle(x, y, size=10, color='blue', alpha=0.8, legend_label='Data Points')
p.line(x, y, line_width=2, color='red', legend_label='Line')
p.legend.location = 'top_left'  # Sets the position of the legend
p.legend.title = 'Legend'       # Sets the legend title


4. Plot Background and Gridlines:-
we can customize the plot's background color, gridlines, and other visual elements using various attributes of the figure object.

example - 

p.background_fill_color = "lightgray"    
p.grid.grid_line_color = "white"         
p.grid.grid_line_alpha = 0.5            
```

```python
Q5.
ANS-

To embed a Bokeh plot into a web page or dashboard using Flask or Django.

we can follow some general steps:-

1.Install the necessary packages:-

Flask: pip install flask
Django: pip install django

2.Create a new Flask or Django project and navigate to the project directory.

3.Copy the Bokeh JavaScript and CSS files:-
Bokeh provides JavaScript and CSS files that need to be included in your web page to render the Bokeh plot correctly.
Copy these files from the Bokeh library installation directory to your project's static file directory. The files you need are bokeh.min.js, bokeh-widgets.min.js, and bokeh.min.css. For example, in Flask, you can copy the files to the static folder, while in Django, you can copy them to the static directory within your app.

4.Create a Flask or Django view function that generates the Bokeh plot:-

In Flask, we can define a route and a view function that generates the Bokeh plot.

example-

from flask import Flask, render_template
from bokeh.plotting import figure, output_file, save

app = Flask(__name__)

@app.route('/')
def index():
    # Create the Bokeh plot
    p = figure()
    p.circle([1, 2, 3], [4, 5, 6])

    # Save the plot to an HTML file
    output_file('bokeh_plot.html')
    save(p)

    # Render the template with the plot
    return render_template('index.html')

if __name__ == '__main__':
    app.run()

    
5. Create an HTML template that includes the Bokeh plot:-

Create an HTML template file (index.html in this example) that includes the Bokeh JavaScript and CSS files, as well as a placeholder for the Bokeh plot. 

Example-

<!DOCTYPE html>
<html>
<head>
    <title>Bokeh Plot</title>
    <link href="/static/bokeh.min.css" rel="stylesheet" type="text/css">
    <script src="/static/bokeh.min.js"></script>
</head>
<body>
    <h1>Bokeh Plot</h1>
    <div id="bokeh_plot">
        {% if bokeh_plot %}
            {% include bokeh_plot %}
        {% endif %}
    </div>
</body>
</html>


6. Start the Flask or Django development server and access the plot in the browser:-

In Flask, run flask run and open http://localhost:5000 in the browser.
In Django, run python manage.py runserver



```
