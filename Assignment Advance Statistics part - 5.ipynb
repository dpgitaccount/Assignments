```python
# Q1.
# ANS-

import scipy.stats as stats

mean = 50
std_dev = 5

sample_size = 12

# standard error
standard_error = std_dev / (sample_size ** 0.5)

# margin of error
margin_of_error = stats.t.ppf(0.975, df=sample_size - 1) * standard_error

# confidence interval
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

print("95% Confidence Interval: ({:.2f}, {:.2f})".format(lower_bound, upper_bound))
```

```python
# Q2.
# ANS-

import scipy.stats as stat
import numpy as np

observed_data = np.array([30, 25, 22, 13, 11, 19]) 

expected_data = np.array([0.2, 0.2, 0.2, 0.1, 0.1, 0.2]) * np.sum(observed_data)

# chi-square goodness of fit test
chi2_stat,p_value = stat.chisquare(observed_data, expected_data)

print("Chi-square statistic: {:.4f}".format(chi2_stat))
print("p-value: {:.4f}".format(p_value))

# Check if the null hypothesis is rejected or not
alpha = 0.05 # Significance llevel

if p_value <= alpha:
    print("The distribution of colors does not match the expected distribution.")
else:
    print("The distribution of colors matches the expected distribution.")
```

```python
# Q6.
# ANS-

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Degrees of freedom
df = 10

# x-values for the plot
x = np.linspace(0, 30, 500)

# chi-square distribution for the given degrees of freedom
chi2_dist = stats.chi2(df)

# Ploting of chi-square
plt.plot(x, chi2_dist.pdf(x), label=f"df = {df}")

# Shade the area corresponding to a chi-square statistic of 15
x_fill = np.linspace(15, 30, 500)
plt.fill_between(x_fill, chi2_dist.pdf(x_fill), color='gray', alpha=0.5)
                 
plt.xlabel("Chi-square statistic")
plt.ylabel("Probability density function")
plt.legend()


plt.show()
```

```python
# Q8.
# ANS- 

import scipy.stats as stats
import numpy as np

# observed data
observed_data = np.array([45, 55])

# Expected data (for a fair coin)
expected_data = np.array([50, 50])

# Calculation of chi-square statistic and p-value
chi2, p_value = stats.chisquare(observed_data, expected_data)

print("Chi-square statistic:", chi2)
print("p-value:", p_value)

alpha = 0.05
if p_value <= alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject null hypothesis")
```

```python
# Q11.
# ANS-
To conduct a hypothesis test to determine if the population mean is significantly different from 70, we can use a one-sample t-test. Here are the steps to perform the hypothesis test:

Step 1: State the null hypothesis (H0) and the alternative hypothesis (Ha):

Null hypothesis (H0): The population mean (μ) is equal to 70.
Alternative hypothesis (Ha): The population mean (μ) is significantly different from 70.
Step 2: Set the significance level (α):
The significance level (α) is given as 0.05 or 5%. This represents the maximum probability of rejecting the null hypothesis when it is actually true.

Step 3: Compute the test statistic:
The test statistic for a one-sample t-test is given by:
t = (sample mean - population mean) / (sample standard deviation / sqrt(sample size))

In this case, the sample mean is 72, the population mean is 70, the sample standard deviation is 10, and the sample size is 30. Plugging in these values, we can calculate the test statistic:
t = (72 - 70) / (10 / sqrt(30))

Step 4: Determine the critical value:
Since we are using a significance level of 0.05 and the test is two-tailed (to detect differences in both directions), we need to find the critical t-values that correspond to an area of 0.025 in each tail of the t-distribution. With 30 degrees of freedom (sample size - 1), the critical t-values can be obtained from a t-table or using statistical software. For simplicity, let's assume the critical t-values are approximately ±2.042.

Step 5: Make a decision:
If the absolute value of the test statistic is greater than the critical value, we reject the null hypothesis; otherwise, we fail to reject the null hypothesis.

Comparing the calculated test statistic from Step 3 with the critical value from Step 4, if |t| > 2.042, we reject the null hypothesis; otherwise, we fail to reject the null hypothesis.

Step 6: Draw a conclusion:
If we reject the null hypothesis, we can conclude that there is sufficient evidence to suggest that the population mean is significantly different from 70. If we fail to reject the null hypothesis, we do not have enough evidence to conclude that the population mean is significantly different from 70.
```
